```python
#Personnalisez vos graphiques avec Matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

prets=pd.read_csv('./utils/pret_immobilier.csv')

data=prets.groupby('ville')['remboursement'].sum()
data=data.reset_index()

#Modifiez les éléments extérieurs
plt.figure(figsize=(10,6))
plt.bar(data['ville'],data['remboursement'])
plt.grid(axis='y')
plt.ylim(0,120000) ##lim horizental
plt.yticks([0,30000,60000,90000,120000])
ca=data['remboursement'].tolist()
for i in range(6):
    plt.text(i-0.2,ca[i],round(ca[i],1))
plt.show()
```
