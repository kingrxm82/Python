```python
# Fusionner des données avec Pandas

#Fusionner des données avecmerge
##merge (La fonction ou méthode prévue par Pandas pour faire une jointure)
##pd.merge(A,B)//A.merge(B)

###pd.merge(A, B, on='id') (lorsque la clé porte le même nom dans chaque data frame)
###C.merge(A, left_on='identifiant', right_on='id')(lorsque la clé n’a pas le même nom)

import pandas as pd

clients = pd.read_csv('./utils/clients.csv')
prets = pd.read_csv('./utils/pret_immobilier.csv')

pd.merge(clients,prets,on='identifiant') #Fusionner les deux dataframes clients et prets dans un seul dataframe data
```

```python
#Explorez les différents types de jointure
import pandas as pd

taille = pd.read_csv('./utils/taille.csv')
poids = pd.read_csv('./utils/poids.csv')

## La jointure interne (est le type de jointure par défaut)
pd.merge(taille,poids,on='id',how='inner') #chaque clé identifiée à droite ou à gauche, elle vérifie s'il y a une correspondance dans l’autre table.

## La jointure à gauche  (est une jointure qui se concentre sur les identifiants de la table située à gauche )
## C'est-à-dire qu’elle conserve forcément toutes les clés qui se trouvent dans la première table
pd.merge(taille,poids,on='id',how='left')

## La jointure à droite (est une jointure qui se concentre sur les identifiants de la table située à drpote )
## C'est-à-dire qu’elle conserve forcément toutes les clés qui se trouvent dans la deuxieme table
pd.merge(taille,poids,on='id',how='right')

## La jointure externe (est une sorte de conjonction des jointures à gauche et à droite)
## C'est-à-dire qu’on conserve TOUTES les clés trouvées, que ce soit à gauche ou à droite
pd.merge(taille,poids,on='id',how='outer')
```
