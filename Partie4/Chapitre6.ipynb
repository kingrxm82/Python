```python
# Agréger des données avec Pandas

#Agréger plusieurs lignes
import pandas as pd
data=pd.read_csv('./utils/pret_immobilier.csv')
##group by
data.groupby('ville').sum() #regroupement par ville et calculer la somme
data.groupby(['ville','type']).sum() #regroupement par ville et type et calculer la somme
data.groupby(['ville','type'])['remboursement'].sum() #regroupement par ville  et type et calculer la somme de rempoubresement par ville et type
data.groupby('ville').agg({'remboursement': ['sum', 'mean']}) # regroupement par ville et afficher la somme et la moyenne de remboursement par ville
```

```python
#Agrégez des lignes et des colonnes
import pandas as pd
data=pd.read_csv('./utils/pret_immobilier.csv')

#methode pivot
##La méthode qui vous permettra de faire cela est la méthode .pivot_table . Celle-ci prend 4 arguments en paramètres :
###index : variable(s) placée(s) en ligne ;
###columns : variable(s) placée en colonne(s) ;
###values : variable sur laquelle on va appliquer la fonction d’agrégation ;
###aggfunc : fonction d’agrégation.
data.pivot_table(index='ville',columns='type',values='remboursement',aggfunc='sum')
```

```python
#Exercice
#https://openclassrooms.com/fr/courses/7771531-decouvrez-les-librairies-python-pour-la-data-science/7857733-agregez-des-donnees-avec-pandas
import numpy as np
import pandas as pd
data=pd.read_csv('./utils/pret_immobilier.csv')

# traitement réalisés précédemment
data = pd.read_csv('./utils/pret_immobilier.csv')
# calcul du taux d'endettement
data['taux_endettement'] = round(data['remboursement'] * 100 / data['revenu'], 2)
# renommer taux en taux_interet
data.rename(columns={'taux':'taux_interet'}, inplace=True)
# calculer le cout total du pret
data['cout_total'] = data['remboursement'] * data['duree']
# calculer les bénéfices mensuels réalisés
data['benefices'] = round((data['cout_total'] * data['taux_interet']/100)/(24), 2)
# création d'une variable risque
data['risque'] = 'Non'
data.loc[data['taux_endettement'] > 35, 'risque'] = 'Oui'

profil_clients=data.groupby('identifiant')[['remboursement','taux_endettement','cout_total','benefices']].sum()
profil_clients.reset_index(inplace=True)
profil_clients.head()

nombre_clients_risuqe = profil_clients.loc[profil_clients['taux_endettement']>35].shape[0]
print('Le nombre des clients qui ont une situtation bancaire risque sont' , nombre_clients_risuqe)

data.groupby(['ville','type'])['benefices'].sum()

data.pivot_table(index='ville',columns='type',values='benefices',aggfunc=np.mean)
```
