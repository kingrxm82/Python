```python
# Filtrer les données du data frame

#Appliquer la sélection via des indices
import pandas as pd
import numpy as np
data = pd.read_csv('./utils/clients.csv')
#Methode iloc (mon_dataframe.iloc[indice_ligne, indice_colonne])
data.iloc[0,2] #Sélectionner le premier ligne  (premier client) (0) et le troisieme colone (nom)(3)
data.iloc[:10,2] #Sélectionner les noms des 10 premiers clients
data.iloc[:10,3] #Sélectionner le genre des 10 premiers clients
data.iloc[:10,:] #Sélectionner tous les colones des 10 premiers clients
data.iloc[:10,1:3] #Sélectionner les noms et les émails des 10 premiers clients
data.iloc[-10:,[1,3]] #Sélectionner l'email et le genre de 10 dérniers clients
#La méthodeiloc peut être utile dans certains cas, mais elle est finalement assez limitée. En effet, elle n’est pas adaptée si on souhaite sélectionner selon une certaine condition 
```

```python
#Utiliser la sélection via des conditions
import pandas as pd
import numpy as np
data = pd.read_csv('./utils/pret_immobilier.csv')
#Méthode loc (mon_dataframe.loc[ condition sur les lignes, colonne(s) ])
data.loc[data['type'] == 'automobile',:] #Sélectionner que les lignes de type automobile
data.loc[data['type']=='automobile','ville'] #Sélectionner que les colomes (ville) ainsi que le type est autombille
data.loc[data['type']=='automobile',['ville','identifiant']] #Sélectionner que les colomes (ville,identifiant) ainsi que le type est autombille
data.loc[data['revenu']>4000,:] #Sélectionner que les lignes ayant un valeur de revenu supérieur à 4000
data.loc[(data['type'] == 'automobile') & (data['revenu']>4000),:] #Sélectionner que les lignes de type automobile avec un valeur de revenu supérieur à 4000
```

```python
#Exercice 
#https://openclassrooms.com/fr/courses/7771531-decouvrez-les-librairies-python-pour-la-data-science/7857549-filtrez-les-donnees-du-data-frame

import numpy as np
import pandas as pd

data = pd.read_csv('./utils/pret_immobilier.csv')
# calcul du taux d'endettement
data['taux_endettement']=round(data['remboursement']*100/data['revenu'],2)
# renommer taux en taux_interet
data.rename(columns={'taux':'taux_interet'},inplace=True)
# calculer le cout total du pret
data['cout_total'] = data['remboursement']*data['duree']
# calculer les bénéfices mensuels réalisés
data['benefices'] = round((data['cout_total']*data['taux_interet']/100)/(24),2)

nombre_clients_risque = data.loc[data['taux_endettement']>35,:].shape[0]
print('Il y a',nombre_clients_risque,'client qui ont dépassé le seuil autorisé')

nombre_clients_risque_paris = data.loc[(data['taux_endettement']>30)&(data['ville']=='PARIS'),:].shape[0]
print('Il y a',nombre_clients_risque_paris,"client qui ont dépassé le seuil autorisé à l'agence de paris")

data['risque']="Non"
data.loc[data['taux_endettement']>35,'risque']="Oui"

data_auto = data.loc[data['type']=='automobile',:]
print('Nous avons accordé',data_auto.shape[0],'prets automobils,dont le cout total moyen est de',data_auto['cout_total'].mean(),'euro')

data_benefices_toulouse = data.loc[data['ville']=='TOULOUSE','benefices'].sum()
print("Le benefice mensuel total réalise par l'agence de toulse est",round(data_benefices_toulouse,2),'euro')
```
