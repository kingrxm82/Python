```python
# Tracez des graphiques avec Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('./utils/pret_immobilier.csv')
data.head()
```

```python
#Les nuages de points(scatter)
plt.scatter(data['revenu'],data['taux'])
```

```python
#Diagramme circulaire (pie)
data_x = data.groupby('ville')['remboursement'].sum()
data_x = data_x.reset_index()

plt.pie(x=data_x['remboursement'],labels=data_x['ville'])
```

```python
#Diagramme  barres (bar)
# plt.bar(height=data_x['remboursement'], x=data_x['ville'])
data_sorted = data.sort_values('remboursement', ascending=False)
plt.bar(height=data_sorted['remboursement'], x=data_sorted['ville'])
```

```python
#Histogramme (hist)
plt.hist(data['revenu'])
```
